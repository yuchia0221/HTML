```python
from svmutil import *
import pandas as pd
import numpy as np
from math import pow
```

```python
y, x = svm_read_problem("satimage.scale")
```

```python
X = pd.DataFrame(x)
X = X.fillna(0)
```

```python
y_pb15 = []
for target in y:
    if target == 3:
        y_pb15.append(1)
    else:
        y_pb15.append(-1)
```

```python
model = svm_train(y_pb15, x, "-s 0 -t 0 -c 10")
```

```python
svm_type = model.get_svm_type()
nr_class = model.get_nr_class()
svr_probability = model.get_svr_probability()
class_labels = model.get_labels()
sv_indices = model.get_sv_indices()
nr_sv = model.get_nr_sv()
is_prob_model = model.is_probability_model()
support_vector_coefficients = model.get_sv_coef()
support_vectors = model.get_SV()
```

```python
test_y, test_x = svm_read_problem("satimage.scale.t")
```

```python
y_pb18 = []
for target in y:
    if target == 6:
        y_pb18.append(1)
    else:
        y_pb18.append(-1)      

y_pb18t = []
for target in test_y:
    if target == 6:
        y_pb18t.append(1)
    else:
        y_pb18t.append(-1)

for i in range(-2, 3, 1):
    cost = pow(10, i)
    print(cost)
    model = svm_train(y_pb18, x, f'-s 0 -t 2 -g 10 -c {cost}')
    result = svm_predict(y_pb18t, test_x, model)
```

```python
for i in range(-1, 4, 1):
    gamma = pow(10, i)
    print(gamma)
    model = svm_train(y_pb18, x, f'-s 0 -t 2 -g {gamma} -c 0.1')
    result = svm_predict(y_pb18t, test_x, model)
```
